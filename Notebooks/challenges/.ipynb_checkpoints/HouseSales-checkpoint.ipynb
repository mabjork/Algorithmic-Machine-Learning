## Feature engineering

```python

houseDataCleanedDF['EncodedStreet'] = pd.get_dummies(houseDataCleanedDF.Street, drop_first=True)
houseDataCleanedDF['EncodedAlley'] = pd.get_dummies(houseDataCleanedDF.Alley, drop_first=True)
houseDataCleanedDF['EncodedUtilities'] = pd.get_dummies(houseDataCleanedDF.Utilities, drop_first=True)

num_neigbourhoods = len(houseDataCleanedDF["Neighborhood"].unique())
houseDataCleanedDF[["N"+str(i) for i in range(num_neigbourhoods-1)]] = pd.get_dummies(houseDataCleanedDF.Neighborhood, drop_first=True)


num_styles = len(houseDataCleanedDF["HouseStyle"].unique())
houseDataCleanedDF[["HS"+str(i) for i in range(num_styles-1)]] = pd.get_dummies(houseDataCleanedDF.HouseStyle, drop_first=True)



#houseDataCleanedDF = houseDataCleanedDF.drop(["Street"],axis=1)
houseDataCleanedDF['EncodedCondition'] = houseDataCleanedDF.SaleCondition\
    .apply(lambda x : 1 if x == "Partial" else 0)

houseDataCleanedDF['EncodedLotConfig'] = houseDataCleanedDF.LotConfig\
    .apply(lambda x : 1 if x == "CulDSac" else 0)
    
houseDataCleanedDF['EncodedBldgType'] = houseDataCleanedDF.BldgType\
    .apply(lambda x : 1 if (x == "1Fam" or x == "TwnhsE")  else 0)
    
houseDataCleanedDF['EncodedMSZoning'] = houseDataCleanedDF.MSZoning\
    .apply(lambda x : 1 if (x == "FV" or x == "RL")  else 0)
    
houseDataCleanedDF['EncodedGarageType'] = houseDataCleanedDF.GarageType\
    .apply(lambda x : 1 if (x == "Attchd" or x == "BuiltIn")  else 0)

houseDataCleanedDF['EncodedGarageType'] = houseDataCleanedDF.GarageType\
    .apply(lambda x : 1 if (x == "Attchd" or x == "BuiltIn")  else 0)
    
houseDataCleanedDF['EncodedSurroundings'] = [1 if x=='PosN' or x=='PosA' or y=='PosN' or y=='PosA' else 0\
    for x,y in zip(houseDataCleanedDF['Condition1'],houseDataCleanedDF['Condition2'])]

houseDataCleanedDF['EncodedSound'] = [1 if x=='Feedr' or x=='Artery' or x=='RRAe' or y=='Feedr'\
    or y=='Artery' or y=='RRAe' else 0 for x,y in zip(houseDataCleanedDF['Condition1'],houseDataCleanedDF['Condition2'])]


#Add feature for centrality
    
```

```python
data = houseDataCleanedDF.select_dtypes(include=[np.number]).interpolate().dropna()
```

```python
y = np.log(data.SalePrice)
X = data.drop(["SalePrice"],axis=1)
features = ['GrLivArea','LotArea','OverallQual','OverallCond','GarageArea','YearBuilt','GarageYrBlt','YearRemodAdd']
#X = X[features]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, random_state=42, test_size=.2)
```

## Linear Regression

```python
lr = LinearRegression()
linearModel = lr.fit(X_train,y_train)
print ("R^2 is:", linearModel.score(X_test, y_test))
```

```python
predictions = linearModel.predict(X_test)

print ('MSLE is:', mean_squared_log_error(y_test, predictions))
```

```python
scores = cross_val_score(linearModel, X_test, y_test, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python

actual_values = y_test
plt.figure(figsize=(15,10))
plt.scatter(predictions, actual_values, alpha=.75,
            color='b') 

plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()
```

### Improving Gradient Boosting

```python
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None, n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    plt.figure(figsize=(15,5))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve( estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.legend(loc="best")
    return plt
```

```python
def GradientBooster(param_grid, n_jobs): 
    estimator = GradientBoostingRegressor()
    cv = ShuffleSplit(X_train.shape[0], n_iter=10, test_size=0.2)
    classifier = GridSearchCV(estimator=estimator, cv=cv, param_grid=param_grid, n_jobs=n_jobs)
    classifier.fit(X_train, y_train)
    print("Best Estimator learned through GridSearch")
    print(classifier.best_estimator_)
    return cv, classifier.best_estimator_
```

```python
param_grid={'n_estimators':[100],
            'learning_rate': [0.1],
            'max_depth':[6],
            'min_samples_leaf':[3],
            'max_features':[1.0],
           }
n_jobs=4
cv,best_est=GradientBooster(param_grid, n_jobs)
```

```python
estimator = GradientBoostingRegressor(n_estimators=best_est.n_estimators, max_depth=best_est.max_depth,
            learning_rate=best_est.learning_rate, min_samples_leaf=best_est.min_samples_leaf,
            max_features=best_est.max_features)
title = "Learning Curves (Gradient Boosted Regression Trees)"
plot_learning_curve(estimator, title, X_train, y_train, cv=cv, n_jobs=n_jobs)
```

```python
estimator.fit(X_train, y_train)
predictions = estimator.predict(X_test)
print ('MSLE is:', mean_squared_log_error(y_test, predictions))
```
